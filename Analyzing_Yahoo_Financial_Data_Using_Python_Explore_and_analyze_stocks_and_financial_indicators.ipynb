```python
project_name = "Analyzing Yahoo Financial Data Using Python: Explore and analyze stocks and financial indicators"
# Project Title
print("=======================================================================================================================")
print("Welcome to the project", project_name)
print("=======================================================================================================================")
print()
# Code comments
print("Data is being analyzed for a project", project_name)
print("Please wait a momentâ€¦")
print()
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
# Mount Google Drive to access the data file
from google.colab import drive
drive.mount('/content/drive')
# Read the CSV file using pandas
path = '/content/drive/My Drive/'
A = pd.read_csv(path + "AI.csv")
# Display the data
print(A)
print("=============================================================================")
# Convert the 'Date' column to datetime format
A['Date'] = pd.to_datetime(A['Date'])
print(A)
print("=============================================================================")
# Calculate the Moving Average
A['Moving Average'] = A['Close'].rolling(window=7).mean()
print(A)
print("=============================================================================")
# Calculate the Price Change
A['Price Change'] = A['Close'].diff()
print(A)
print("=============================================================================")
# Calculate the Short-term Moving Average
A['Short-term Moving Average'] = A['Close'].rolling(window=7).mean()
print(A)
print("=============================================================================")
# Calculate the Long-term Moving Average
A['Long-term Moving Average'] = A['Close'].rolling(window=30).mean()
print(A)
print("=============================================================================")
# Calculate the Slope
A['Slope'] = A['Short-term Moving Average'].diff() / A['Long-term Moving Average'].diff()
print(A)
print("=============================================================================")
# Calculate the Return
A['Return'] = A['Close'].pct_change()
print(A)
print("=============================================================================")
# Calculate the Risk
A['Risk'] = A['Return'].rolling(window=7).std()
print(A)
print("=============================================================================")
# Plot the Stock Price
plt.plot(A['Date'], A['Close'])
plt.xlabel('Date')
plt.ylabel('Price')
plt.title('Stock Price')
plt.show()
print("=============================================================================")
# Plot the Stock Price with Moving Average
plt.plot(A['Date'], A['Close'], label='Price')
plt.plot(A['Date'], A['Moving Average'], label='Moving Average')
plt.xlabel('Date')
plt.ylabel('Price')
plt.title('Stock Price with Moving Average')
plt.legend()
plt.show()
print("=============================================================================")
# Calculate Simple Moving Averages (SMA)
df = pd.DataFrame(A)
df['SMA_10'] = df['Close'].rolling(window=10).mean()
df['SMA_20'] = df['Close'].rolling(window=20).mean()
df['Signal'] = 0
df.loc[df['SMA_10'] > df['SMA_20'], 'Signal'] = 1  # Signal for buying
df.loc[df['SMA_10'] < df['SMA_20'], 'Signal'] = -1  # Signal for selling
print(df)
print("=============================================================================")
# Define a function to calculate SMA
def calculate_sma(data, window):
    return data['Close'].rolling(window=window).mean()
print("=============================================================================")
# Calculate SMA with window size 5 and 10
A['SMA_5'] = calculate_sma(A, 5)
A['SMA_10'] = calculate_sma(A, 10)
print("=============================================================================")
# Plot Closing Price and SMA
plt.plot(A['Date'], A['Close'], label='Closing Price')
plt.plot(A['Date'], A['SMA_5'], label='SMA 5')
plt.plot(A['Date'], A['SMA_10'], label='SMA 10')
plt.xlabel('Date')
plt.ylabel('Price')
plt.title('Closing Price and SMA')
plt.legend()
plt.xticks(rotation=45)
plt.show()
print("=============================================================================")
# Define a function to calculate the slope
def calculate_slope(data):
    return np.polyfit(range(len(data)), data, 1)[0]
print("=============================================================================")
# Calculate the slope of Closing Price with a window of 10
A['Slope_10'] = A['Close'].rolling(window=10).apply(calculate_slope)
print(A)
# Plot the slope
plt.plot(A['Date'], A['Slope_10'])
plt.xlabel('Date')
plt.ylabel('Slope')
plt.title('Slope of Closing Price')
plt.xticks(rotation=45)
plt.show()
print("=============================================================================")
# Calculate mean, max, min, and volume
mean_price = A['Close'].mean()
max_price = A['High'].max()
min_price = A['Low'].min()
volume = A['Volume'].sum()
print("=============================================================================")
# Print the results
print('Mean Closing Price:', mean_price)
print('Highest Price:', max_price)
print('Lowest Price:', min_price)
print('Total Volume:', volume)
print("=============================================================================")
```
